## 2. Run Validation Tests

````python
VERDICTS = ["STRONG BUY", "INTERESTED", "WATCHLIST", "PASS", "HARD PASS"]

# The Phase 2 analysis template has 5 possible verdicts:
# STRONG BUY / INTERESTED / WATCHLIST / PASS / HARD PASS
#
# For validation, we are not trying to predict the exact intensity.
# We are checking whether the direction matches historical reality.
# - Known positions should be positive: STRONG BUY or INTERESTED
# - Known passes should be negative: PASS or HARD PASS
# - WATCHLIST is neutral, so it is not counted as correct by default.
BUY_VERDICTS  = {"STRONG BUY", "INTERESTED"}
PASS_VERDICTS = {"PASS", "HARD PASS"}
NEUTRAL_VERDICTS = {"WATCHLIST"}

COUNT_WATCHLIST_AS_CORRECT_FOR_POSITIONS = False
COUNT_WATCHLIST_AS_CORRECT_FOR_PASSES = False

if COUNT_WATCHLIST_AS_CORRECT_FOR_POSITIONS:
    BUY_VERDICTS.add("WATCHLIST")
if COUNT_WATCHLIST_AS_CORRECT_FOR_PASSES:
    PASS_VERDICTS.add("WATCHLIST")


def normalize_verdict(value) -> str:
    """
    Converts model output into exactly one of:
    STRONG BUY, INTERESTED, WATCHLIST, PASS, HARD PASS, ERROR
    """
    if value is None:
        return "ERROR"

    s = str(value).strip().upper()
    s = s.replace("_", " ").replace("-", " ")
    s = re.sub(r"[^A-Z\[\]X ]+", " ", s)
    s = re.sub(r"\s+", " ", s).strip()

    # First: detect explicit checked boxes from the Phase 2 template.
    checkbox_patterns = [
        (r"\[\s*X\s*\]\s*STRONG\s+BUY", "STRONG BUY"),
        (r"\[\s*X\s*\]\s*INTERESTED", "INTERESTED"),
        (r"\[\s*X\s*\]\s*WATCH\s*LIST", "WATCHLIST"),
        (r"\[\s*X\s*\]\s*WATCHLIST", "WATCHLIST"),
        (r"\[\s*X\s*\]\s*HARD\s+PASS", "HARD PASS"),
        (r"\[\s*X\s*\]\s*PASS", "PASS"),
    ]
    for pattern, label in checkbox_patterns:
        if re.search(pattern, s):
            return label

    # Second: detect exact labels, most specific first.
    if "HARD PASS" in s:
        return "HARD PASS"
    if "STRONG BUY" in s:
        return "STRONG BUY"
    if "INTERESTED" in s:
        return "INTERESTED"
    if "WATCHLIST" in s or "WATCH LIST" in s:
        return "WATCHLIST"
    if "PASS" in s:
        return "PASS"

    # Third: map simplified outputs.
    if s in {"BUY", "LIKELY BUY", "YES", "PASS INITIAL SCREEN", "INITIAL SCREEN YES"}:
        return "INTERESTED"
    if s in {"NOT BUY", "NO BUY", "AVOID", "REJECT", "NO", "INITIAL SCREEN NO"}:
        return "PASS"

    if "PASS" in s:
        return "PASS"
    if "BUY" in s:
        return "INTERESTED"

    return "ERROR"


def extract_verdict_from_text(raw: str) -> str:
    """
    Extracts a verdict from either:
    1. JSON value fields, or
    2. Phase 2 markdown/checklist format.
    """
    if not raw:
        return "ERROR"

    verdict = normalize_verdict(raw)
    if verdict != "ERROR":
        return verdict

    # Try a more targeted search around "Probable Conclusion" if present.
    m = re.search(r"Probable Conclusion:(.*?)(?:Position Sizing|Catalyst|Confidence|$)", raw, flags=re.IGNORECASE | re.DOTALL)
    if m:
        verdict = normalize_verdict(m.group(1))
        if verdict != "ERROR":
            return verdict

    return "ERROR"


def is_correct_verdict(test_type: str, verdict: str) -> bool:
    verdict = normalize_verdict(verdict)
    if test_type == "position":
        return verdict in BUY_VERDICTS
    if test_type == "pass":
        return verdict in PASS_VERDICTS
    return False


def compact_evidence(evidence_obj, max_items_per_section: int = 3):
    """
    Keeps evidence small and robust even if the JSON structure changes.
    """
    if isinstance(evidence_obj, dict):
        compact = {}
        for k, v in evidence_obj.items():
            if isinstance(v, list):
                compact[k] = v[:max_items_per_section]
            elif isinstance(v, dict):
                compact[k] = dict(list(v.items())[:max_items_per_section])
            else:
                compact[k] = v
        return compact
    if isinstance(evidence_obj, list):
        return evidence_obj[:10]
    return evidence_obj


def parse_json_response(raw: str) -> dict:
    """
    Robustly parses Gemini output.
    If Gemini returns the full Phase 2 markdown/checklist instead of JSON,
    this still extracts the checked verdict instead of returning ERROR.
    """
    raw = (raw or "").strip()
    cleaned = re.sub(r"^```json\s*", "", raw, flags=re.IGNORECASE)
    cleaned = re.sub(r"^```\s*", "", cleaned)
    cleaned = re.sub(r"\s*```$", "", cleaned)

    try:
        parsed = json.loads(cleaned)
        parsed["verdict"] = normalize_verdict(parsed.get("verdict", "ERROR"))
        return parsed
    except Exception:
        pass

    # Fallback: extract the first JSON object from the response.
    match = re.search(r"\{.*\}", cleaned, flags=re.DOTALL)
    if match:
        try:
            parsed = json.loads(match.group(0))
            parsed["verdict"] = normalize_verdict(parsed.get("verdict", "ERROR"))
            return parsed
        except Exception:
            pass

    # Final fallback: parse the full text/checklist response.
    return {
        "verdict": extract_verdict_from_text(raw),
        "reasoning": raw[:500],
        "initial_screen": "",
        "key_positives": [],
        "key_concerns": []
    }


VALIDATION_PROMPT_TEMPLATE = """
You are applying {investor}'s investment framework to evaluate: {company} ({ticker}).

Context on {company}: {reason}

Based ONLY on the investor DNA and evidence below, respond with a JSON object:
{{
  "verdict": "<one of: STRONG BUY | INTERESTED | WATCHLIST | PASS | HARD PASS>",
  "reasoning": "<2-3 sentence explanation in the investor's voice>",
  "initial_screen": "<YES | MAYBE | NO>",
  "key_positives": ["<str>", "<str>"],
  "key_concerns": ["<str>", "<str>"]
}}

Rules:
- Return ONLY valid JSON. No markdown, no preamble.
- Use exactly one of these verdict labels: STRONG BUY, INTERESTED, WATCHLIST, PASS, HARD PASS.
- Do not output the full Phase 2 checklist format.
- For known positions, do not simply copy the context; evaluate whether the DNA logically supports ownership.
- For known passes, do not simply copy the context; evaluate whether the DNA logically supports passing.
- Do not invent quotes.

INVESTOR DNA:
{dna}

EVIDENCE (key sections):
{evidence}
"""


def run_validation_test(company: str, ticker: str, reason: str) -> dict:
    evidence_snippet = json.dumps(
        compact_evidence(evidence, max_items_per_section=3),
        indent=2,
        ensure_ascii=False
    )

    prompt = VALIDATION_PROMPT_TEMPLATE.format(
        investor=INVESTOR_DISPLAY_NAME,
        company=company,
        ticker=ticker,
        reason=reason,
        dna=investor_dna[:6000],
        evidence=evidence_snippet
    )

    response = client.models.generate_content(
        model=MODEL_NAME,
        contents=prompt,
        config=types.GenerateContentConfig(
            temperature=0.1,
            max_output_tokens=1000,
            response_mime_type="application/json"
        )
    )

    parsed = parse_json_response(response.text)
    parsed["raw_verdict"] = parsed.get("verdict", "ERROR")
    parsed["verdict"] = normalize_verdict(parsed.get("verdict", "ERROR"))
    parsed["raw_response"] = response.text[:1000]

    return parsed


results = []

print(f"Testing {len(KNOWN_POSITIONS)} known positions...")
for pos in KNOWN_POSITIONS:
    print(f"  {pos['company']}...", end=" ")
    result = run_validation_test(pos["company"], pos["ticker"], pos["reason"])
    verdict = normalize_verdict(result.get("verdict", "ERROR"))
    correct = is_correct_verdict("position", verdict)

    results.append({
        "type": "position",
        "company": pos["company"],
        "ticker": pos["ticker"],
        "expected_verdicts": ", ".join(["STRONG BUY", "INTERESTED"]),
        "actual": "BUY",
        "predicted_verdict": verdict,
        "raw_verdict": result.get("raw_verdict", ""),
        "correct": correct,
        "reasoning": result.get("reasoning", ""),
        "positives": "; ".join(result.get("key_positives", [])),
        "concerns": "; ".join(result.get("key_concerns", []))
    })
    print(f"{'✅' if correct else '❌'} {verdict}")
    time.sleep(1)


print(f"\nTesting {len(KNOWN_PASSES)} known passes...")
for pas in KNOWN_PASSES:
    print(f"  {pas['company']}...", end=" ")
    result = run_validation_test(pas["company"], pas["ticker"], pas["reason"])
    verdict = normalize_verdict(result.get("verdict", "ERROR"))
    correct = is_correct_verdict("pass", verdict)

    results.append({
        "type": "pass",
        "company": pas["company"],
        "ticker": pas["ticker"],
        "expected_verdicts": ", ".join(["PASS", "HARD PASS"]),
        "actual": "PASS",
        "predicted_verdict": verdict,
        "raw_verdict": result.get("raw_verdict", ""),
        "correct": correct,
        "reasoning": result.get("reasoning", ""),
        "positives": "; ".join(result.get("key_positives", [])),
        "concerns": "; ".join(result.get("key_concerns", []))
    })
    print(f"{'✅' if correct else '❌'} {verdict}")
    time.sleep(1)

print("\n✅ All tests complete")
````

## 3. Accuracy Report

````python
df = pd.DataFrame(results)

total        = len(df)
correct      = int(df["correct"].sum())
accuracy     = correct / total * 100 if total else 0

pos_df       = df[df["type"] == "position"]
pass_df      = df[df["type"] == "pass"]
pos_acc      = pos_df["correct"].mean() * 100 if len(pos_df) else 0
pass_acc     = pass_df["correct"].mean() * 100 if len(pass_df) else 0

print("═" * 75)
print(f"  VALIDATION REPORT — {INVESTOR_DISPLAY_NAME}")
print("═" * 75)
print(f"  Overall Accuracy  : {correct}/{total} ({accuracy:.0f}%)")
print(f"  Known Positions   : {int(pos_df['correct'].sum())}/{len(pos_df)} ({pos_acc:.0f}%)")
print(f"  Known Passes      : {int(pass_df['correct'].sum())}/{len(pass_df)} ({pass_acc:.0f}%)")
print("─" * 75)

print("\nDetailed Results:")
display_cols = [
    "type",
    "company",
    "ticker",
    "expected_verdicts",
    "predicted_verdict",
    "correct"
]
print(df[display_cols].to_string(index=False))

print("\nMisclassifications:")
misses = df[~df["correct"]]
if misses.empty:
    print("  None — perfect score!")
else:
    for _, row in misses.iterrows():
        print(f"  ❌ {row['company']} | Expected one of [{row['expected_verdicts']}] → Got {row['predicted_verdict']}")
        print(f"     {row['reasoning']}")
````

## 4. Save Results

````python
val_dir = BASE_DIR / "Validation"
val_dir.mkdir(parents=True, exist_ok=True)

csv_path = val_dir / f"{INVESTOR_NAME}_validation_results.csv"
df.to_csv(csv_path, index=False)

summary = {
    "investor": INVESTOR_DISPLAY_NAME,
    "date": str(date.today()),
    "overall_accuracy": round(accuracy, 1),
    "position_accuracy": round(pos_acc, 1),
    "pass_accuracy": round(pass_acc, 1),
    "total_tests": total,
    "correct": int(correct),
    "misclassifications": misses[["company", "actual", "predicted_verdict", "reasoning"]].to_dict(orient="records")
}

json_path = val_dir / f"{INVESTOR_NAME}_validation_summary.json"
json_path.write_text(json.dumps(summary, indent=2), encoding="utf-8")

print(f"✅ Saved to {val_dir}")
print(f"   {csv_path.name}")
print(f"   {json_path.name}")
````
